```python
import ipywidgets as widgets
from IPython.display import display

# Display screen
screen = widgets.Text(
    value='',
    placeholder='0',
    description='',
    disabled=False,
    layout=widgets.Layout(width='300px')
)

# Button click function
def on_button_click(b):
    if b.description == "=":
        try:
            screen.value = str(eval(screen.value))
        except:
            screen.value = "Error"
    elif b.description == "C":
        screen.value = ""
    else:
        screen.value += b.description

# Button names
buttons = [
    '7', '8', '9', '/',
    '4', '5', '6', '*',
    '1', '2', '3', '-',
    '0', '.', '=', '+',
    'C'
]

# Create buttons
button_widgets = []
for btn in buttons:
    button = widgets.Button(
        description=btn,
        layout=widgets.Layout(width='70px', height='40px')
    )
    button.on_click(on_button_click)
    button_widgets.append(button)

# Arrange buttons in grid
grid = widgets.GridBox(
    button_widgets,
    layout=widgets.Layout(
        grid_template_columns="repeat(4, 75px)"
    )
)

# Display calculator
display(screen)
display(grid)
```
