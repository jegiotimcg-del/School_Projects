```python
# Snake Game in Python (Text-based)
import random
import os
import time

board_size = 5
snake = [[2,2]]
food = [random.randint(0,board_size-1), random.randint(0,board_size-1)]
score = 0

def print_board():
    for i in range(board_size):
        for j in range(board_size):
            if [i,j] == snake[-1]:
                print("S", end=" ")
            elif [i,j] in snake:
                print("o", end=" ")
            elif [i,j] == food:
                print("F", end=" ")
            else:
                print(".", end=" ")
        print()
    print(f"Score: {score}\n")

while True:
    print_board()
    move = input("Move (W/A/S/D): ").upper()
    head = snake[-1].copy()
    if move == "W": head[0] -= 1
    if move == "S": head[0] += 1
    if move == "A": head[1] -= 1
    if move == "D": head[1] += 1

    # Check collision
    if head in snake or not (0 <= head[0] < board_size) or not (0 <= head[1] < board_size):
        print("Game Over!")
        break

    snake.append(head)

    if head == food:
        score += 1
        food = [random.randint(0,board_size-1), random.randint(0,board_size-1)]
    else:
        snake.pop(0)

    time.sleep(0.2)
    os.system('clear')
```
