```python
import ipywidgets as widgets
from IPython.display import display
```

```python
name = widgets.Text(description="Name:")
roll = widgets.Text(description="Roll No:")

math = widgets.IntText(description="Math:")
english = widgets.IntText(description="English:")
sciology = widgets.IntText(description="Sciology:")
urdu = widgets.IntText(description="Urdu:")
computer = widgets.IntText(description="Computer:")
education=widgets.IntText(description="Education:")

```

```python
btn = widgets.Button(description="Calculate Result",
                     button_style="success")
```

```python
output = widgets.Output()
```

```python
def calculate(b):
    with output:
        output.clear_output()

        total = math.value + english.value + sciology.value + urdu.value + computer.value + education.value
        percentage = (total / 600) * 100

        if percentage >= 80:
            grade = "A+"
        elif percentage >= 70:
            grade = "A"
        elif percentage >= 60:
            grade = "B"
        elif percentage >= 50:
            grade = "C"
        else:
            grade = "Fail"

        print("Student Name:", name.value)
        print("Roll Number:", roll.value)
        print("Total Marks:", total, "/ 600")
        print("Percentage:", round(percentage, 2), "%")
        print("Grade:", grade)
```

```python
btn.on_click(calculate)
```

```python
display(name, roll, math, english, sciology, urdu, computer,education, btn, output)
```
